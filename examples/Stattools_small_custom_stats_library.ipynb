```python
#this imports the quantile calculation
from stattools import quantileCalc
```

```python
#example with a single Gaussian:
import numpy as np
mean = 0
sigma = 10
x = np.arange(-10*np.pi,10*np.pi,0.01)
y = np.exp(-0.5*((x - mean)/sigma)**2)
qc = quantileCalc(x, y)
ylvl, intervals = qc.getquantilevertical()
#printing the y level:
print ylvl
#printing the intervals:
print intervals
```

```python
#making the ploty
qc.plot()
```
